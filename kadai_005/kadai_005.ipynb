クラス勉強　イベント開催：インデント勉強

```python
#インデントなし
from datetime import datetime
class Event:
  #インデント2
  def __init__(self, name: str, date: str, location: str):
      #インデント4
      """コンストラクタ
      :param name: イベント名
      :param date: 開催日('YYYY-MM-DD')
      :param location: 開催場所
      """

      self.name = name
      self.date = datetime.strptime(date, "%Y-%m-%d")
      self.location = location

  #インデント2 開催日までの日数を出力
  def days_until_event(self) -> int:
      #インデント4 開催日までの日数を出力
      """開催日までの日数を返す"""
      today = datetime.today()
      return (self.date - today).days

#===== ここからが実行される =====
#インデントなし
event = Event("プログラミング初心者向けセミナー", "2026-02-15", "東京国際フォーラム")

#インデントなし
#各属性の値を出力
print(f"イベント名: {event.name}")
print(f"開催日: {event.date.strftime('%Y-%m-%d')}")
print(f"開催場所: {event.location}")
print(f"開催日までの日数: {event.days_until_event()}日")
```

##kadai_イベント開催：正解パターン

```python
from datetime import datetime

class Event:
    def __init__(self, name: str, date: str, location: str):
        """コンストラクタ
        :param name: イベント名
        :param date: 開催日('YYYY-MM-DD')
        :param location: 開催場所
        """
        self.name = name
        self.date = datetime.strptime(date, "%Y-%m-%d")
        self.location = location

# 開催日までの計算を定義
    def days_until_event(self) -> int:
        """開催日までの日数を返す"""
        today = datetime.today()
        return (self.date - today).days


# ===== ここからが実行される =====
event = Event(
    "プログラミング初心者向けセミナー",
    "2026-02-15",
    "東京国際フォーラム"
)
#各属性の値を出力
print(f"イベント名: {event.name}")
print(f"開催日: {event.date.strftime('%Y-%m-%d')}")
print(f"開催場所: {event.location}")
#開催日までの日数を出力
print(f"開催日までの日数: {event.days_until_event()}日")
```

##kadai_017

```python
#名前(name)と年齢(age)の属性を持つHumanクラスを作成
#Humanクラスは、以下の条件で標準出力(print)するcheck_adultメソッドを追加
#ageが20以上の場合に大人、そうでない場合に大人でない
#Humanクラスのインスタンスを複数生成してリストに追加し、check_adultメソッドを呼び出し
#リスト・ディクショナリ、条件分岐、繰り返し処理、関数、変数のスコープ、クラスで課題の仕様通りのプログラム作成

class Human:
    def __init__(self, name: str, age: int):
        """Humanクラスのコンストラクタ
        :param name:
        :param age: 年齢
        """
        self.name = name
        self.age = age

    def check_adult(self):
        """大人かどうかを判定して出力する"""
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は大人ではありません。")


# ===== ここから実行部分 =====
# Humanクラスのインスタンスを複数生成してリストに追加
humans = [
    Human("ぶり", 30),
    Human("うの", 18),
    Human("うに", 20),
    Human("らく", 15)
]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
